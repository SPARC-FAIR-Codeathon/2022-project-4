```python
#Import libraries
from ipyleaflet import Map, VectorTileLayer, TileLayer
```

```python
tile_urls= ['https://mapcore-demo.org/current/flatmap/v2/flatmap/83f6c97d571b67fb4c273e20287b53b4f0a1f70780d3d6a2a282e66cef5f9473/tiles/vagus_test_image/{z}/{x}/{y}',
           'https://mapcore-demo.org/current/flatmap/v2/flatmap/83f6c97d571b67fb4c273e20287b53b4f0a1f70780d3d6a2a282e66cef5f9473/mvtiles/{z}/{x}/{y}']

m = Map(
    basemap=TileLayer(
                    url=tile_urls[0], 
                    min_zoom=2),
    min_zoom=2,
    max_zoom=10,
    zoom=5,
    scroll_wheel_zoom=True,
    dragging=True,
    attribution_control=False,
    zoom_snap=False,
)


vl = VectorTileLayer(url=tile_urls[1])
m.add_layer(vl)
m
```
